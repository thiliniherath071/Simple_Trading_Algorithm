```python
list = [1,2,3,4,5,6]
window_size = 3
  
i = 0
# Initialize an empty list to store moving averages
moving_averages = []
  
# Loop through the array to consider
# every window of size 3
while i < len(list) - window_size + 1:
    
    # Store elements from i to i+window_size
    # in list to get the current window
    window = list[i : i + window_size]
  
    # Calculate the average of current window
    window_average = round(sum(window) / window_size, 2)
      
    # Store the average of current
    # window in moving average list
    moving_averages.append(window_average)
      
    # Shift window to right by one position
    i += 1
  
print(moving_averages)
```

```python
df = data['Adj Close'].to_frame()
df['SMA30'] = df['Adj Close'].rolling(100).mean()
df.dropna(inplace=True)
df.head(45)
```

```python
df[['Adj Close', 'SMA30']].plot(figsize=(16,10))
plt.show()
```

```python
# Create a column for signal (1 for buy, -1 for sell)
df['Signal'] = 0
df.loc[df['SMA30'] > df['Adj Close'], 'Signal'] = 1
df.loc[df['SMA30'] < df['Adj Close'], 'Signal'] = -1

# Create a column for position (1 for long, -1 for short)
df['Position'] = df['Signal'].diff()


res = df[df.Position.isin([2.0, -2.0])]
res
```

```python
res['abc'] = res['Adj Close'].diff()
res.dropna(inplace=True)
res
```
